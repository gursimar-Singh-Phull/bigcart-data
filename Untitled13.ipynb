```python
 
# Assuming 'df' is your DataFrame with 'Product Name', 'Sales', and 'Category' columns
filtered_df = df[df['Category'] == 'Office Supplies']  # Filter for 'Office Supplies' category

# Extract product names and sales values
product_names = filtered_df['Product Name']
sales_values = filtered_df['Sales']

# Create the bar chart with a larger figure size
plt.figure(figsize=(16, 8))

# Use a more vibrant color for the bars with more spacing
plt.bar(range(1, len(sales_values) + 1), sales_values, color='#69b3a2', edgecolor='black', width=0.8)

# Add labels and title with enhanced fonts
plt.xlabel('Product Name', fontsize=14, fontweight='bold')
plt.ylabel('Sales Values', fontsize=14, fontweight='bold')
plt.title('Sales for Office Supplies', fontsize=16, fontweight='bold')

# Customize x-axis labels with product names, show every second label to reduce congestion
plt.xticks(range(1, len(sales_values) + 1, 2), product_names[::2], rotation=45, ha='right', fontsize=12)

# Set the y-axis ticks to a smaller range, with better spacing
plt.yticks(range(0, max(sales_values) + 1, 100))  # Adjust the step for y-axis ticks to reduce congestion

# Display sales values on top of the bars with a cleaner appearance
for i, value in enumerate(sales_values):
    plt.text(i + 1, value + 10, str(value), ha='center', fontsize=12, fontweight='bold', color='black')

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Set tight layout to avoid overlapping labels and improve spacing
plt.tight_layout()

# Show the plot
plt.show()

```
